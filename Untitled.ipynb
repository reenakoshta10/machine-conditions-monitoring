```python
import librosa as lb
import librosa.display
from glob import glob
import os 
import matplotlib.pyplot as plt

import numpy as np
import pandas as pd
import csv
```

```python
#create an csv file 
def create_csv():

    header = ['filename', 'class']

    with open('dude1.csv', 'w') as f:
        writer = csv.writer(f)
        writer.writerow(header)
        for name in glob('**/*.wav', recursive=True):
            data = []
            arr = name.split('\\')
            data.append(name)
            data.append(arr[3])
            writer.writerow(data)


```

```python
create_csv()
```

```python
df = pd.read_csv('dude1.csv')
df
```

```python
#stacking the mffcs
def mffc_array(i:int,data:object)->object:
    filename = df['filename'][i]
    librosa_load = lb.load(str(filename),sr=None)
    audio, sample_rate = librosa_load
    mfcc = lb.feature.mfcc(audio,sr=sample_rate,n_mfcc=1)
    mfcc = mfcc.flatten()
    if len(data) ==0:
        data = mfcc
    else:
        data = np.vstack((data,mfcc))
    
    return data
```

```python
#getting the targets y
def get_target():
    lst = []
    for i in df.index:
        if df['class'][i] == 'abnormal':
            lst.append(0)
        else:
            lst.append(1)
    y =np.array(lst)
    y = np.reshape(y,(df.index.stop,-1))
    return y
```

```python
#hstacking the features and the targets
def fuse_target_mffc(data:object,y:object)->object:
    data = np.hstack((data, y))
    return data
```

```python
#looping through csv file to calculate the mffc from every file
def manageMffc()->object:
    header = ['X','y']
    lst=[]
    data =np.array(lst)
    with open('mffc2.csv','w')as f:
        writer = csv.writer(f)
        writer.writerow(header)
        for i in df.index:
            data = mffc_array(i,data)
    return data
```

```python
data = fuse_target_mffc(manageMffc(),get_target())
data
```

```python
#ultimate df
dff = pd.DataFrame(data)
dff
```

```python
#preprocessing dff -> mffc dataset; the last column is the y. 
#0 == abnormal and 1 equals normal

X = dff.iloc[:,0:-1].values
y= dff.iloc[:,-1].values
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.2)
```

```python
scaler = StandardScaler()
scaler.fit(X_train)
X_train = scaler.transform(X_train)
X_test = scaler.transform(X_test)
```

```python
classifier = KNeighborsClassifier(n_neighbors=100)
classifier.fit(X_train,y_train)
```

```python
y_predict = classifier.predict(X_test)
```

```python
print(classification_report(y_test, y_predict))
print(confusion_matrix(y_test,y_predict))
```

```python
acc = accuracy_score(y_test,y_predict)
acc
```

```python
ks = 200
mean_acc = np.zeros((ks))

for i in range(1,ks+1):
    model = KNeighborsClassifier(n_neighbors=i)
    model.fit(X_train,y_train)
    y_predict1 = model.predict(X_test)
    mean_acc[i-1]= accuracy_score(y_test, y_predict1)
```
