```python
import librosa as lb
import numpy as np
```

```python
y, sr = lb.load("data/-6_dB_slider/slider/id_00/normal/00000001.wav")
```

```python
tempo, beat_frames = lb.beat.beat_track(y=y, sr=sr)
```

```python
print('Estimated tempo: {:.2f} beats per minute'.format(tempo))
```
